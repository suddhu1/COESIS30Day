## 3. Box Plor (Numerical - Categorical )

```python
sns.boxplot(data=dataframe, x='Sex', y='Age')
```

```python
sns.boxplot(data=dataframe, x='Sex', y='Age',hue='Survived')
```

## 4. Displot (Numerical - Categorical )

```python
sns.displot(data=dataframe, x='Age', kde=True)
```

```python
sns.kdeplot(data=dataframe[dataframe['Survived'] == 0]['Age'], label='Not Survived', fill =True)
sns.kdeplot(data=dataframe[dataframe['Survived'] == 1]['Age'], label='Survived', fill =True)
plt.xlabel("Age")
plt.ylabel("Density")
plt.title("KDE Plot of Age by Survival")
plt.legend()
plt.show()
```

## 5. HeatMap (Categorical - Categorical )

```python
dataframe.head()

```

```python
pd.crosstab(dataframe['Pclass'], dataframe['Survived'])
```

```python
sns.heatmap(pd.crosstab(dataframe['Pclass'], dataframe['Survived']))
```

```python
dataframe.groupby('Pclass')['Survived'].value_counts().unstack().plot(kind='bar', stacked=True)
```

```python
dataframe.groupby('Sex')['Survived'].value_counts().unstack().plot(kind='bar', stacked=True)
```

## 6.ClusterMap (Categorical - Categorical )

```python
pd.crosstab(dataframe['SibSp'], dataframe['Survived'])
```

```python
sns.clustermap(pd.crosstab(dataframe['SibSp'], dataframe['Survived']))
```
