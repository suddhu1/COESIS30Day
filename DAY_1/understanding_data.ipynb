```python
# Histogram
# defination: A histogram is a graphical representation of the distribution of numerical data.
#  It is an estimate of the probability distribution of a continuous variable (quantitative variable) and was first introduced by Karl Pearson.
#  It is a kind of bar graph.
import matplotlib.pyplot as plt 
plt.hist(df['Age'], bins=50, color='green')
```

```python
# Displot 
# defination: A displot is a plot that combines the matplotlib hist
#  function with the seaborn kdeplot() and rugplot() functions.
#  It is used to visualize the distribution of a dataset.
sns.displot(df['Age'], kde=True)
```

```python
sns.displot(df['Age'], kde=True, stat='probability')
```
