## 2. K-1 OneHotEncoding 

```python
pd.get_dummies(df1,columns=['fuel','owner'],dtype=int,drop_first=True) # will remove the first column that is formed from the categories  
```

## OneHotEncoding with the Top Categories 

```python
counts = df1['brand'].value_counts()
```

```python
df1['brand'].nunique()
threshold = 100
```

```python
repl = counts[counts <= threshold].index
```

```python
pd.get_dummies(df1['brand'].replace(repl, 'uncommon'),dtype=int).sample(5)
```
